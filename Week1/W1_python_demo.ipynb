```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option("display.max_columns", None)
```

```python
df_houseprice = pd.read_csv("house_price.csv")
df_houseprice
```

## Choose Col

```python
df_simple = df_houseprice[['SalePrice','GarageArea','GarageCond']]
df_simple
```

```python
df_simple.GarageCond.value_counts()
```

## Data Viz

```python
plt.figure(figsize=(6,6))
sns.scatterplot("GarageArea", "SalePrice", data=df_simple)
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot("GarageArea", "SalePrice", data=df_simple, hue="GarageCond")
```
